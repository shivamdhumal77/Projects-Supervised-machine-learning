```python
#splitting the dependent and independent variables
X = df.drop(['Class'], axis= 1)
y = df['Class']
print(X.shape)
print(y.shape)
# getting just the values for the sake of processing
```

```python
#Training and Testing Data Bifurcation

#We will be dividing the dataset into two main groups. 
#One for training the model and the other for Testing our trained model’s performance.

from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test  = train_test_split(X,y , test_size=0.6, random_state=42)
```

```python
#Code : Building a Random Forest Model using scikit learn
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
# Evlauating the classifier
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, matthews_corrcoef
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
n_outliers = len(fraud)
n_errors   = (y_pred != y_test).sum()
print("the model is Random Forest classifier")

acc = accuracy_score(y_test, y_test)
print("The accuracy is Random Forest CLassifier")

prec = precision_score(y_test, y_pred)
print("the accuracy is {}".format(prec))

rec  = recall_score(y_test, y_pred)
print("The F1-score is {}".format(rec))

f1  = f1_score(y_test, y_pred)
print("the f1 socre is {}".format(f1))

MCC = matthews_corrcoef(y_test, y_pred)
print("The Mathews correlation coefficient is {}".format(MCC))
```
