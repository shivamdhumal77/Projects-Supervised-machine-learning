```python
df['Spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train , y_test = train_test_split(df.Message, df.Spam, test_size=0.6)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# The CountVectorizer from sklearn.feature_extraction.text is a tool in scikit-learn that converts a collection of text documents into a matrix of token counts. It is a common text preprocessing step in natural language processing (NLP) and machine learning when working with textual data.

#Here's what it does:

#Key Functions:
#Tokenization: It splits the text into individual words or tokens. By default, it separates tokens by whitespace and removes punctuation.

#Vocabulary Building: It builds a vocabulary of all unique words found in the corpus (the collection of text documents).

#Counting Word Occurrences: For each document, it counts how many times each word (from the vocabulary) appears.

#Output:
#The output is a sparse matrix where:
#Rows correspond to documents.
#Columns correspond to terms (unique words) from the vocabulary.
#Values in the matrix represent the count of each word in the respective document.
from sklearn.naive_bayes  import MultinomialNB
from sklearn.pipeline import Pipeline
clf = Pipeline(steps=[('vectorizer', CountVectorizer()), ('nb', MultinomialNB())])

# TRAINING THE MODEL
```

```python
#Here now the prediction
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred  = clf.predict(X_test)
accuracy_score = accuracy_score(y_test, y_pred)
print(accuracy_score)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
```
